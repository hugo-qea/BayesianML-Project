### Pima

```python
X_pima_1 = pd.read_csv("../data/pima/data1/x.csv")
Y_pima_1 = pd.read_csv("../data/pima/data1/y.csv")
```

```python
X_pima_1.head().T
```

```python
Y_pima_1.head().T
```

```python
print(Y_pima_1.describe().T)
print("="*50)
print(Y_pima_1.diabetes.value_counts())
```

```python
pima_data = pd.concat([X_pima_1, Y_pima_1], axis=1)
correlation_matrix = pima_data.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.show()
```

```python
descriptive_stats = X_pima_1.describe()
positive_Y = X_pima_1[Y_pima_1['diabetes'] == 1]
negative_Y = X_pima_1[Y_pima_1['diabetes'] == 0]
positive_stats = positive_Y.describe()
negative_stats = negative_Y.describe()
```

```python
positive_stats
```

```python
negative_stats
```

```python
plt.figure(figsize=(20, 20))
for i, column in enumerate(X_pima_1.columns):
    plt.subplot(3, 3, i+1)
    sns.boxplot(x=Y_pima_1['diabetes'], y=X_pima_1[column])
    plt.title(f"{column}")
plt.tight_layout()
plt.show()
```
